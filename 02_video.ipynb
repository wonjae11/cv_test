## 02. 비디오 입출력
- 비디오의 재생 : 연속된 이미지를 재생하는 것
- 프레임 : 연속된 이미지 중 하나
- 비디오를 재생하기 위해서는 프레임을 연속적으로 재생


### 2-1. 비디오 파일 입출력
- 'cv2.VideoCapture("파일경로")' : 경로의 파일을 볼러옴
- 'VideoCapture.isOpened()' : 파일이 열렸는지 확인
- 'VideoCapture.read()' : 파일을 읽어서 frame을 반환
    - ret : 프레임이 있는지 체크, frame : 불러온 이미지
    - 'VideoCapture.release()' : 영상 입력 종료

```python
import cv2
```

```python
cat_video_path = "videos/cat.mp4"
```

```python
cap = cv2.VideoCapture(cat_video_path)

while cap.isOpened():
    ret, frame = cap.read()
    print(ret, frame)
    if not ret:
        break

    cv2.imshow("Video", frame)

    if cv2.waitKey(30) == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

### 2-2. 영상의 프레임 불러오기
- 'VideoCapture.get(cv2.CAP_PROP_FPS)' 

```python
cap = cv2.VideoCapture(cat_video_path)
fps = cap.get(cv2.CAP_PROP_FPS)

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    cv2.imshow("Video", frame)

    if cv2.waitKey(int(1000/fps)) == ord("q"):
        break

cap.release
cv2.destroyAllWindows()
```

### 2-3. 카메라 입출력

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("카메라가 없어요")
    exit()

while True:
    ret, frame = cap.read()

    if not ret:
        print("불러올 이미지가 없어요")
        break

    cv2.imshow("My Camera", frame)

    if cv2.waitKey(1) == ord("q"):
        print("사용자 입력에 의해 종료되었어요.")
        break

cap.release()
cv2.destroyAllWindows()
```

### 2-4. 비디오 쓰기
- 'cv2.VideoWriter(파일경로, fourcc, fps, size)' : 영상을 쓰기


```python
print(*"DIVX")
```

```python
cap = cv2.VideoCapture(cat_video_path)

# VideoWriter 생성성
fourcc = cv2.VideoWriter_fourcc(*"H264")
fps = cap.get(cv2.CAP_PROP_FPS)
width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter("output/cat.mp4", fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    out.write(frame) # 소리는 들어가지 X
    cv2.imshow("Video", frame)

    if cv2.waitKey(int(1000/fps)) == ord("q"):
        break

out.release()
cap.release()
cv2.destroyAllWindows()
```
