## 09. 마우스 이벤트

```python
import cv2
img = cv2.imread("images/winter.jpg")
name = "Mouse Event"
cv2.namedWindow(name)

def on_mouse(event, x, y, flags, _):
    if event == cv2.EVENT_LBUTTONDOWN:
        copied = img.copy()
        cv2.circle(copied, (x, y), 50, (255,255,0), cv2.FILLED)
        cv2.imshow(name, copied)


cv2.setMouseCallback(name, on_mouse)
cv2.imshow(name, img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread("images/desk.jpg")
original_img = img.copy()
point_list = []
mouse_pos = None
width, height = 400, 300

def on_mouse(event, x, y, flags, _):
    global point_list, mouse_pos, img
    if event == cv2.EVENT_LBUTTONDOWN:
        if len(point_list) < 4:
            point_list.append((x, y))
            lines_img()

        if len(point_list) == 4:
            warp_document()

    elif event == cv2.EVENT_MOUSEMOVE:
        mouse_pos = (x, y)
        if len(point_list) > 0 and len(point_list) < 4:
            lines_img()


def lines_img():
    global img, point_list, mouse_pos
    img = original_img.copy()

    for point in point_list:
        cv2.circle(img, point, 5, (0, 255, 255), cv2.FILLED)

    for i in range(1, len(point_list)):
        cv2.line(img, point_list[i - 1], point_list[i], (0, 255, 0), 2)

    if len(point_list) > 0 and mouse_pos is not None:
        cv2.line(img, point_list[-1], mouse_pos, (0, 0, 255), 2)

    if len(point_list) == 4:
        cv2.line(img, point_list[-1], point_list[0], (0, 255, 0), 2)

    cv2.imshow("img", img)

def warp_document():
    global point_list
    pts = np.float32(point_list)
    dst = np.array([[0, 0], [width, 0], [width, height], [0, height]], dtype=np.float32)
    mat = cv2.getPerspectiveTransform(pts, dst)
    result = cv2.warpPerspective(original_img, mat, (width, height))
    cv2.imshow("Warped", result)

cv2.imshow("img", img)
cv2.setMouseCallback("img", on_mouse)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread("images/card.jpg")
point_list = []

# 최종 결과를 출력하는 함수입니다
def show_result():
  width, height = 600, 350
  src = np.array(point_list, dtype=np.float32)
  # src = np.float32(point_list)
  target = np.array([[0,0], [width, 0], [width, height], [0, height]], dtype=np.float32)
  mat = cv2.getPerspectiveTransform(src, target)
  result = cv2.warpPerspective(img, mat, (width, height))
  cv2.imshow("result", result)

# 마우스 이벤트와 함께 실행되는 콜백함수입니다.
drawing = False
def on_mouse(event, x, y, flags, _):
  global drawing
  coppied = img.copy()
  if event == cv2.EVENT_LBUTTONDOWN:
    drawing = True
    point_list.append((x,y))

  if drawing:
    prev_point = None
    for point in point_list:
      cv2.circle(coppied, point, 10, (0,0,255), cv2.FILLED)
      if prev_point:
        cv2.line(coppied, prev_point, point, (0,0,255), 5)
      prev_point = point

    next_point = (x,y)
    if len(point_list) == 4:
      next_point = point_list[0]
      show_result()
    cv2.line(coppied, point_list[-1], next_point, (0,0,255), 5)

  cv2.imshow(name, coppied)

# 기본적인 코드 실행입니다.
name = "Scanner"
cv2.namedWindow(name)
cv2.setMouseCallback(name, on_mouse)
cv2.imshow(name, img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
