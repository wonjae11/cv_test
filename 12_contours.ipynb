## 12. 윤곽선 검출
- 경계선을 연결한 선
- 흑백 이미지 -> 이진화 -> 윤곽선 찾기(findCoutours) -> 윤곽선 그리기(drawContours)

```python
import cv2

WINTER = "images/winter.jpg"
```

### 12-1. 윤곽선 검출

```python
img = cv2.imread(WINTER)
copied = img.copy()
copied_2 = img.copy()


# 그레이스케일로 변환
gray = cv2.cvtColor(copied, cv2.COLOR_BGR2GRAY)

# 이진화
ret, binary = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

# 윤곽선 찾기
contours, hierachy = cv2.findContours(binary, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
contours_2, hierachy_2 = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)


# 윤곽선 그리기
cv2.drawContours(copied, contours, -1, (0,255,0), 2)
cv2.drawContours(copied_2, contours_2, -1, (0,255,0), 2)


cv2.imshow("img", img)
cv2.imshow("contours",copied)
cv2.imshow("contours_2",copied_2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 12-2. 윤곽선 경계 사각형

```python
img = cv2.imread(WINTER)
copied = img.copy()

gray = cv2.cvtColor(copied, cv2.COLOR_BGR2GRAY)

ret, binary = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

contours, hierachy = cv2.findContours(binary, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

for contour in contours:
    x, y, width, height = cv2.boundingRect(contour)
    cv2.rectangle(copied, (x, y), (x+width, y+height), (255,0,255), 2)
cv2.imshow("contours",copied)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("images/vehicles.png")
copied = img.copy()
name = "Vehicles"
cv2.namedWindow(name)

# 윤곽선을 찾고 : 그레이스케일 -> 이진화 -> findContours
gray = cv2.cvtColor(copied, cv2.COLOR_BGR2GRAY)
ret, binary = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
contours, hierachy = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 윤곽선의 인덱스를 선택해서 -> 트랙바이용
cv2.createTrackbar("Index", name, 0, len(contours)-1, lambda x:x)

while True:
    target = img.copy()
    index = cv2.getTrackbarPos("Index", name)

    # 해당 인덱스의 bounding rect를 추출하고 -> cv2.boundingRect
    x, y, width, height = cv2.boundingRect(contours[index])
    # 화면에 사각형을 그린다 -> cv2.rectangle
    cv2.rectangle(target, (x,y), (x+width, y+height), (0,255,0),2)
    cv2.imshow(name, target)

    if cv2.waitKey(1) == ord("q"):
        break

cv2.destroyAllWindows()
```

```python
img = cv2.imread("images/vehicles.png")
copied = img.copy()
name = "Vehicles"
cv2.namedWindow(name)

gray = cv2.cvtColor(copied, cv2.COLOR_BGR2GRAY)
ret, binary = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
contours, hierachy = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 윤곽선의 면적이 특정 크기 이상인 윤곽선을 새로운 리스트에 저장
MIN_AREA = 1000
filtered_contours = []
for contour in contours:
    if cv2.contourArea(contour) > MIN_AREA:
        filtered_contours.append(contour)

cv2.createTrackbar("Index", name, 0, len(filtered_contours)-1, lambda x:x)

while True:
    target = img.copy()
    index = cv2.getTrackbarPos("Index", name)
    print(cv2.contourArea(filtered_contours[index]))

    x, y, width, height = cv2.boundingRect(filtered_contours[index])

    cv2.rectangle(target, (x,y), (x+width, y+height), (0,255,0),2)
    cv2.imshow(name, target)

    if cv2.waitKey(1) == ord("q"):
        break

cv2.destroyAllWindows()
```

### 실습. 카드 하나씩 새 창에 표시

```python
PLAYING_CARD = "images/playing_cards.png"

img = cv2.imread(PLAYING_CARD)
copied = img.copy()
name = "Playing cards"
cv2.namedWindow(name)

gray = cv2.cvtColor(copied, cv2.COLOR_BGR2GRAY)
ret, binary = cv2.threshold(gray, -1, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
contours, hierachy = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

MIN_AREA = 25000

filtered_contours = []
for contour in contours:
    if cv2.contourArea(contour) > MIN_AREA:
        filtered_contours.append(contour)

cv2.createTrackbar("Index", name, 0, len(filtered_contours)-1, lambda x:x)

while True:
    target = img.copy()
    index = cv2.getTrackbarPos("Index", name)
    print(cv2.contourArea(filtered_contours[index]))

    x, y, width, height = cv2.boundingRect(filtered_contours[index])

    cv2.rectangle(target, (x,y), (x+width, y+height), (0,255,0),2)
    card = img[y:y + height, x:x + width]
    cv2.imshow(name, target)
    cv2.imshow(f"Card {index + 1}", card)

    if cv2.waitKey(1) == ord("q"):
        break

cv2.destroyAllWindows()


```

```python
import cv2 as cv
img = cv.imread("images/many_cards.png")
img = cv.resize(img, (1200, 600))
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
ret, otsu = cv.threshold(gray, -1, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)
contours, hierachy = cv.findContours(otsu, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_NONE)
filtered_contours = []

for contour in contours:
    if cv.contourArea(contour) > 500:
        filtered_contours.append(contour)

cv.imshow('img', img)
cv.namedWindow('img')
cv.createTrackbar('Index', 'img', 0, len(filtered_contours) - 1, lambda x: x)

window_name = None
previous_index = -1

while True:
    index = cv.getTrackbarPos('Index', 'img')
    
    if index != previous_index:
        if window_name:
            cv.destroyWindow(window_name)
        x1, y1, width, height = cv.boundingRect(filtered_contours[index])
        x2 = x1 + width
        y2 = y1 + height
        separated_img = img[y1:y2, x1:x2]
        separated_img = cv.resize(separated_img, (150, 200))
        cv.imshow(window_name, separated_img)
        previous_index = index
    
    if cv.waitKey(1) != -1:
        break
    
cv.destroyAllWindows()
```
