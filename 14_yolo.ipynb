```python
from ultralytics import YOLO
import cv2
import numpy as np
```

### YOLO 모델 불러오기

```python
model = YOLO('yolo11n.pt')

img = cv2.imread("images/포항윈터.png")

results = model.predict(img, conf=0.5)

annotated_frame = results[0].plot()

cv2.imshow("img", annotated_frame)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 캠화면에 적용

```python
model = YOLO('yolo11n.pt')

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("카메라가 없어요")
    exit()

while True:
    ret, frame = cap.read()

    if not ret:
        print("불러올 이미지가 없어요")
        break

    results = model.predict(frame, conf=0.5)

    annotated_frame = results[0].plot()

    cv2.imshow("My Camera", annotated_frame)

    if cv2.waitKey(1) == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
